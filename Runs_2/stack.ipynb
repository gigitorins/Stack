```python
nsim = 100
filenames = [("run%i.dat" % (i+1)) for i in range(nsim) ]
fig = plt.figure(figsize=(14, 8))

for i, item in enumerate(filenames):
    media, maxval = (np.genfromtxt(item, usecols=k, autostrip=True) for k in range(2))
    plt.plot(media[:-1],label='Fitness media '+str(i+1))

plt.xlabel('Generazione', fontsize = 13)
plt.ylabel('Fitness', fontsize = 13)
plt.title('Fitness media', fontsize = 15)
plt.grid(True)
plt.show()
```

```python
# Caricamento dati da file
fig = plt.figure(figsize=(14, 16))
mediatot, maxvaltot = (np.zeros(200) for i in range(2))
mediatot2, maxvaltot2 = (np.zeros(200) for i in range(2))
counter = [1 for x in range(200)]
genmed = np.genfromtxt("a_generations.dat", usecols=0, autostrip=True)

for item in filenames:
    media, maxval = (np.genfromtxt(item, usecols=k, autostrip=True) for k in range(2))
    for i in range(len(media)-1): counter[i] += 1
    mediatot[:len(media)-1] += media[:-1]
    maxvaltot[:len(maxval)-1] += maxval[:-1]
    mediatot2[:len(media)-1] += media[:-1]
    maxvaltot2[:len(maxval)-1] += maxval[:-1]
    mediatot2[len(media):] += media[-2]
    maxvaltot2[len(media):] += maxval[-2]

mediatot /= counter
maxvaltot /= counter
mediatot2 /= nsim
maxvaltot2 /= nsim
mediavects = [[mediatot, maxvaltot],[mediatot2, maxvaltot2]]
display(Math(r'Generazione \,\, media: \,\, %f \pm %f ' % (np.mean(genmed), np.std(genmed)) ))

for j, item in enumerate(mediavects):
    ax = fig.add_subplot(2,1,j+1)
    ax.plot(item[0][:int(max(genmed)-2)],label='Fitness media')
    ax.plot(item[1][:int(max(genmed)-2)],label='Fitness massima', c="r")
    ax.axvline(np.mean(genmed), c="darkorange", label='Generazione media')
    ax.axvspan(np.mean(genmed)-np.std(genmed),np.mean(genmed)+np.std(genmed),color="bisque")
    ax.set_xlabel('Generazione', fontsize = 15)
    ax.legend(loc=0)  
    ax.grid(True)

#fig.tight_layout()
plt.show()
```
